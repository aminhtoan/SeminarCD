```python
"""
Bai06 - Lab thực hành Chương 12: Tài liệu hóa Mã nguồn với GenAI
Sách: Supercharged Coding with Gen AI

Mục tiêu:
  - Minh họa cách viết docstring theo chuẩn Google Style
  - So sánh code không có / có docstring
  - Sử dụng GenAI (GitHub Copilot / ChatGPT) để tự động sinh docstring

Prompt mẫu dùng với Copilot/ChatGPT:
  "Write a Google-style docstring for the following Python function
   that calculates L1 and L2 distance between two NumPy arrays."
"""

import numpy as np


# ─────────────────────────────────────────────
# BEFORE: Hàm không có docstring (kém chất lượng)
# ─────────────────────────────────────────────
def calculate_distance_no_doc(dist_type, a, b):
    if a.shape != b.shape:
        return None
    if dist_type == "L1":
        return np.sum(np.abs(a - b))
    elif dist_type == "L2":
        return np.sqrt(np.sum((a - b) ** 2))
    else:
        return None


# ─────────────────────────────────────────────
# AFTER: Hàm với docstring Google Style (GenAI hỗ trợ)
# ─────────────────────────────────────────────
def calculate_distance(dist_type: str, a: np.ndarray, b: np.ndarray) -> float | None:
    """Calculate the distance between two NumPy arrays.

    Supports L1 (Manhattan) and L2 (Euclidean) distance metrics.
    Both arrays must have the same shape.

    Args:
        dist_type (str): Distance metric type. Accepted values: 'L1' or 'L2'.
        a (np.ndarray): First input array.
        b (np.ndarray): Second input array, must match shape of `a`.

    Returns:
        float | None: Computed distance value, or None if shapes differ
            or an invalid dist_type is provided.

    Raises:
        TypeError: If `a` or `b` are not NumPy arrays.

    Example:
        >>> import numpy as np
        >>> a = np.array([1.0, 2.0, 3.0])
        >>> b = np.array([4.0, 6.0, 3.0])
        >>> calculate_distance("L1", a, b)
        7.0
        >>> calculate_distance("L2", a, b)
        5.0
    """
    if not isinstance(a, np.ndarray) or not isinstance(b, np.ndarray):
        raise TypeError("Both inputs must be NumPy arrays.")

    if a.shape != b.shape:
        return None

    if dist_type == "L1":
        return float(np.sum(np.abs(a - b)))
    elif dist_type == "L2":
        return float(np.sqrt(np.sum((a - b) ** 2)))
    else:
        return None


# ─────────────────────────────────────────────
# PEP 257 – Single-line docstring (ngắn gọn)
# ─────────────────────────────────────────────
def normalize_vector(v: np.ndarray) -> np.ndarray:
    """Return the unit vector of v (L2-normalized)."""
    norm = np.linalg.norm(v)
    if norm == 0:
        return v
    return v / norm


# ─────────────────────────────────────────────
# Module-level docstring cho class (Google Style)
# ─────────────────────────────────────────────
class DistanceCalculator:
    """A calculator for various distance metrics between vectors.

    This class wraps distance computation methods and keeps track of
    the number of calculations performed.

    Attributes:
        call_count (int): Number of distance calculations performed.

    Example:
        >>> calc = DistanceCalculator()
        >>> a = np.array([0.0, 0.0])
        >>> b = np.array([3.0, 4.0])
        >>> calc.l2(a, b)
        5.0
    """

    def __init__(self):
        """Initialize DistanceCalculator with zero call count."""
        self.call_count = 0

    def l1(self, a: np.ndarray, b: np.ndarray) -> float:
        """Compute Manhattan (L1) distance between two vectors.

        Args:
            a (np.ndarray): First vector.
            b (np.ndarray): Second vector, same shape as `a`.

        Returns:
            float: L1 distance.
        """
        self.call_count += 1
        return float(np.sum(np.abs(a - b)))

    def l2(self, a: np.ndarray, b: np.ndarray) -> float:
        """Compute Euclidean (L2) distance between two vectors.

        Args:
            a (np.ndarray): First vector.
            b (np.ndarray): Second vector, same shape as `a`.

        Returns:
            float: L2 distance.
        """
        self.call_count += 1
        return float(np.sqrt(np.sum((a - b) ** 2)))


# ─────────────────────────────────────────────
# Demo chạy thực tế
# ─────────────────────────────────────────────
if __name__ == "__main__":
    a = np.array([1.0, 2.0, 3.0])
    b = np.array([4.0, 6.0, 3.0])

    print("=" * 50)
    print("Ch12 Lab: Documenting Code with GenAI")
    print("=" * 50)

    print(f"\n[Function] L1 distance: {calculate_distance('L1', a, b)}")
    print(f"[Function] L2 distance: {calculate_distance('L2', a, b)}")
    print(f"[Function] Invalid type: {calculate_distance('L3', a, b)}")

    v = np.array([3.0, 4.0])
    print(f"\n[Normalize] {v} → {normalize_vector(v)}")

    calc = DistanceCalculator()
    print(f"\n[Class] L1: {calc.l1(a, b)}")
    print(f"[Class] L2: {calc.l2(a, b)}")
    print(f"[Class] Total calls: {calc.call_count}")

    # Kiểm tra docstring qua help()
    print("\n[Docstring via help()]:")
    help(calculate_distance)
```

```python
def compute_batch_distances(vectors: list[np.ndarray], metric: str = "L2") -> list[float]:
    """Compute distances between consecutive vectors in a batch.

    Applies the specified distance metric to calculate distances
    between each pair of consecutive vectors in the input list.

    Args:
        vectors (list[np.ndarray]): List of NumPy arrays representing vectors.
            Each array must have the same shape.
        metric (str, optional): Distance metric type. Defaults to "L2".
            Accepted values: "L1" or "L2".

    Returns:
        list[float]: List of computed distances between consecutive vector pairs.
            Returns an empty list if fewer than 2 vectors are provided.

    Raises:
        TypeError: If vectors is not a list or contains non-NumPy array elements.
        ValueError: If vectors have mismatched shapes.

    Example:
        >>> v1 = np.array([0.0, 0.0])
        >>> v2 = np.array([3.0, 4.0])
        >>> v3 = np.array([6.0, 8.0])
        >>> compute_batch_distances([v1, v2, v3], metric="L2")
        [5.0, 5.0]
    """
    if not isinstance(vectors, list):
        raise TypeError("vectors must be a list")
    
    distances = []
    for i in range(len(vectors) - 1):
        distances.append(calculate_distance(metric, vectors[i], vectors[i + 1]))
    return distances
```
