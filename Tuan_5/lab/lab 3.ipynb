```python
def get_geometric_mean_of_two_numbers(
    a: float,
    b: float,
) -> float:
    """Calculate the geometric mean of two numbers.

    The geometric mean of two numbers a and b is defined as the square root of their product:
    geometric_mean = sqrt(a * b)

    Args:
        a: The first number.
        b: The second number.

    Returns:
        The geometric mean of a and b.
    """
    # Part 1: Calculate the product
    product = a * b

    # Part 2: Calculate the square root
    geometric_mean = product ** 0.5

    return geometric_mean

num1: float = 5.0
num2: float = 20.0
print(f"The geometric mean of {num1} and {num2} is: {get_geometric_mean_of_two_numbers(num1, num2)}")
```

Lab 3.2 – introducing completion and keyboard
shortcuts for geometric mean calculations

```python
def get_geometric_mean(*nums: float) -> float:
    """
    Get the geometric mean of a sequence of numbers
    """
    from functools import reduce
    from math import pow
    
    if not nums:
        raise ValueError("Cannot calculate the geometric mean of an empty sequence")
    
    product = reduce(lambda a, b: a * b, nums)
    if product < 0 and len(nums) % 2 == 0:
        raise ValueError("Cannot calculate the geometric mean of negative numbers")
    
    return pow(product, 1 / len(nums))
```
