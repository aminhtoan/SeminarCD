```python
import openai
from openai import OpenAI
import os

api_key = os.getenv("OPENAI_API_KEY")


if __name__ == "__main__":
    client = OpenAI(api_key=api_key)
    
    completion = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {"role": "user", "content": "What is the FizzBuzz problem?"}
        ]
    )
    
    print(completion.choices[0].message.content)
```

```python
if __name__ == "__main__":
    # Tạo client
    client = OpenAI(api_key=api_key)
    
    # Gọi API với các tham số bổ sung
    completion = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "user", "content": "What is the Two Sum problem?"}
        ],
        n=1,              
        temperature=0.2,  
        max_tokens=100   
    )
    
    # In thông tin tokens
    print("Completion Tokens:", completion.usage.completion_tokens)
    print("Total Tokens:", completion.usage.total_tokens)
    
    # In output
    print("\nOutput:")
    print(completion.choices[0].message.content)
```
