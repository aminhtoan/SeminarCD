Multiple service responses with Lab 2.2

```python
import google.generativeai as genai
api_key = "AIzaSyCn02m6xN0MuWArAoRSC0PtBGUMFijfYVY"

try:
    # Configure API
    genai.configure(api_key=api_key)

    generation_config = {
        "temperature": 2,
        "max_output_tokens": 500
    }

    model = genai.GenerativeModel(
        "gemini-2.5-flash",
        generation_config=generation_config,
        system_instruction="You are a hiring manager at a tech company."
    )

    prompt = "Explain the Two Sum problem."

    # Generate 3 responses
    for i in range(3):
        response = model.generate_content(prompt)

        print(f"\nOutput {i+1}:")
        print(response.text)

        if hasattr(response, "usage_metadata"):
            print("Completion Tokens:", response.usage_metadata.candidates_token_count)

except Exception as e:
    print(f"Lỗi: {e}")
```

```python


USER_PROMPT = """\ndef print_fibonacci_sequence(n: int) -> None:\n"""
SYSTEM_PROMPT = "You will be provided with a Python function signature. Your task is to implement the function. Return code only."

def get_code_with_instructions(code: str) -> str:
    """
    Add a comment to the code for specific code completion instruction
    :param code: Python code as string
    :return: The code with additional instruction - 'Complete this code'
    """
    return code + "\n# Complete this code"

try:
    genai.configure(api_key=api_key)
    model = genai.GenerativeModel(
        "gemini-2.5-flash",
        generation_config=generation_config,
        system_instruction=SYSTEM_PROMPT
)
    prompt = get_code_with_instructions(USER_PROMPT)
    response = model.generate_content(prompt)
    print("Suggested code:")
    print(response.text)
except Exception as e:
    print(f"Lỗi: {e}")
```
